**Кейс-задача №1**

Дан одномерный массив А размерности N. Найти сумму отрицательных элементов, расположенных между максимальным и минимальным.

```python
def sum_negatives_between_min_max(arr):
    if not arr:
        return 0

    max_val = max(arr)
    min_val = min(arr)

    max_index = arr.index(max_val)
    min_index = arr.index(min_val)

    start, end = min(max_index, min_index), max(max_index, min_index)

    sum_neg = 0
    for i in range(start + 1, end):
        if arr[i] < 0:
            sum_neg += arr[i]

    return sum_neg

# Пример использования:
A = [3, -2, 5, -1, 2, 5, -3, 4, -6, 2]
result = sum_negatives_between_min_max(A)
print(result)  # Вывод: -1 и -3 находится между 5(максимум) и -6(минимум). Их сумма дает -4.
```
