#**Веб-приложения с Delphi 10.2**

Создание простого веб-приложения с Delphi 10.2, IIS и PostgreSQL
В этом руководстве мы создадим простое веб-приложение для управления списком книг (библиотека) с использованием Delphi 10.2, IIS в качестве веб-сервера и PostgreSQL в качестве базы данных.

##**1. Настройка базы данных PostgreSQL**

Сначала создадим базу данных и таблицу для хранения информации о книгах.

```sql
-- Создаем базу данных
CREATE DATABASE library_db;

-- Подключаемся к базе данных
\c library_db

-- Создаем таблицу книг
CREATE TABLE books (
    id SERIAL PRIMARY KEY,
    title VARCHAR(100) NOT NULL,
    author VARCHAR(100) NOT NULL,
    year_published INTEGER,
    isbn VARCHAR(20),
    created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);
```

##**2. Создание веб-приложения в Delphi 10.2**

Откройте Delphi 10.2 и создайте новый проект:

- File → New → Other → Delphi Projects → WebBroker → Web Server Application
- Выберите "Apache Dynamic Link Module" (для работы с IIS через модуль Apache)
- Или "ISAPI/NSAPI Dynamic Link Library" (для прямого использования с IIS)

Настройте модуль WebModule:
- Добавьте компоненты TFDConnection, TFDQuery, TFDPhysPgDriverLink для работы с PostgreSQL
- Добавьте компоненты TDataSetTableProducer или TPageProducer для генерации HTML


##**3. Настройка IIS для работы с приложением**

Установите IIS (если не установлен):
- Панель управления → Программы → Включение или отключение компонентов Windows
- Включите "Internet Information Services"

Настройте IIS для работы с ISAPI:
- Откройте "Диспетчер служб IIS"
- Создайте новый веб-сайт или используйте "Default Web Site"
- Добавьте виртуальный каталог и укажите путь к вашему DLL-файлу
- В настройках обработчиков добавьте поддержку ISAPI-dll

Настройте права доступа:
- Убедитесь, что пользователь IUSR имеет права на выполнение вашего DLL-файла
- Проверьте права доступа к базе данных PostgreSQL для пользователя, указанного в подключении

##**4. Компиляция и развертывание**
- Скомпилируйте проект в Delphi (получите DLL-файл)
- Скопируйте DLL-файл в каталог на сервере IIS (обычно C:\inetpub\wwwroot\your_app)
- Настройте IIS как описано выше
- Откройте браузер и перейдите по адресу http://localhost/your_app.dll

##**Заключение:**

Вы создали простое веб-приложение на Delphi, которое:
- Подключается к PostgreSQL
- Отображает список книг в виде HTML-таблицы
- Может быть расширено для добавления, редактирования и удаления книг.
